This notebook loads the csv files containing the explanations obtained from the best model using LIME and uses a dataset containing variable names and descriptions to match up the variables to their textual descriptions.

```python
import pandas as pd
```

```python
exp = pd.read_csv('../results/variable_explanations.csv')
```

```python
exp.head()
```

```python
del exp['Unnamed: 0']
```

```python
exp.shape
```

```python
missingness = pd.DataFrame(exp.isnull().sum())
```

```python
freq_dict = {}
c=0
for r in missingness.iterrows():
    freq_dict[r[0]] = 50-r[1][0]
    if 50-r[1][0] > 1:
        c+=1
        print(r[0], 50-r[1][0])
print(c)
```

```python
freq_dict
```

```python
meta.head()
```

```python
descs = {}
for c in list(exp.columns):
    descs[c] = meta.loc[c]['label']
```
